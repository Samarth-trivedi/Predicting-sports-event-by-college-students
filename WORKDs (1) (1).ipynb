```python
import pandas as pd
import re

# Load the CSV file into a DataFrame
file_path = 'C:/Users/samar/Downloads/DS DATA/ReSponse.csv'  # Replace with the correct file path
df = pd.read_csv(file_path)

# Data Cleaning

# Function to convert height to cm
def height_to_cm(height_str):
    if pd.notna(height_str) and isinstance(height_str, str):
        match = re.match(r'(\d+\.?\d*)', height_str)
        if match:
            return float(match.group(1))
    return None

# Function to clean the weight column
def clean_weight(weight_str):
    if pd.notna(weight_str) and isinstance(weight_str, str):
        match = re.match(r'(\d+\.?\d*)', weight_str)
        if match:
            return float(match.group(1))
    return None

# Clean 'What's your height?(Write in cm)' and 'What's your weight?(Write in numbers)' columns
df['What\'s your height?(Write in cm)'] = df['What\'s your height?(Write in cm)'].apply(height_to_cm)
df['What\'s your weight?(Write in numbers)'] = df['What\'s your weight?(Write in numbers)'].apply(clean_weight)

# Remove the 'Score' column
df = df.drop(columns=['Score'])

# Rename the 'Department?' column to 'Department'
df = df.rename(columns={'Department?': 'Department'})

# Display the cleaned data
print("Cleaned Data:")
print(df.head())

# Save the cleaned data to a new CSV file if needed
# df.to_csv('cleaned_data.csv', index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read the CSV file
data = pd.read_csv('C:/Users/samar/Downloads/DS DATA/FINAL DATA.csv')

# Calculate the percentage of students interested in each sport
sport_percentage = data['Sport'].value_counts(normalize=True) * 100

# Count the number of male and female students
gender_counts = data['Gender'].value_counts()

# Determine which branch prefers each sport more
branch_sport_preference = data.groupby(['Sport', 'Department'])['Gender'].count().unstack(fill_value=0)

# Set up the figure and axes
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Plot the percentage of students interested in each sport
sns.barplot(x=sport_percentage.index, y=sport_percentage.values, ax=axes[0, 0])
axes[0, 0].set_title('Percentage of Students Interested in Each Sport')
axes[0, 0].set_xlabel('Sport')
axes[0, 0].set_ylabel('Percentage')

# Plot the number of male and female students
sns.barplot(x=gender_counts.index, y=gender_counts.values, ax=axes[0, 1])
axes[0, 1].set_title('Number of Male and Female Students')
axes[0, 1].set_xlabel('Gender')
axes[0, 1].set_ylabel('Count')

# Plot the branch preferences for each sport
sns.heatmap(branch_sport_preference, annot=True, cmap='YlGnBu', ax=axes[1, 0])
axes[1, 0].set_title('Branch Preferences for Each Sport')
axes[1, 0].set_xlabel('Department')
axes[1, 0].set_ylabel('Sport')

# Rotate the x-labels for better readability
for ax in axes[0]:
    ax.set_xticklabels(ax.get_xticklabels(), rotation=45)

# Adjust the layout
plt.tight_layout()

# Show the plots
plt.show()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, confusion_matrix
from sklearn.impute import SimpleImputer

# Assuming 'data' is your DataFrame containing the provided dataset
data = pd.read_csv('FINAL DATA.csv')

# Drop any rows with missing values in the target variable
data = data.dropna(subset=['Comman hindrance '])

# Encode categorical variables
label_encoder = LabelEncoder()
data['Gender'] = label_encoder.fit_transform(data['Gender'])
data['Sport'] = label_encoder.fit_transform(data['Sport'])
data['Department'] = label_encoder.fit_transform(data['Department'])

# Convert 'How many day a week you will play' to numerical values
data['How many day a week you will play'] = data['How many day a week you will play'].map({
    'Less than 1 hour': 0,
    '1-3 hours': 1,
    'More than 3 hours': 2
})

# Define features (X) and target variable (y)
X = data.drop('Comman hindrance ', axis=1)
y = data['Comman hindrance ']

# Handle missing values
imputer = SimpleImputer(strategy='most_frequent')  # You can also try 'mean' or 'median'
X_imputed = imputer.fit_transform(X)

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_imputed, y, test_size=0.2, random_state=42)

# Create a RandomForestClassifier
model = RandomForestClassifier(random_state=42)

# Train the model
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred, average='weighted')
precision = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')
conf_matrix = confusion_matrix(y_test, y_pred)

print(f"Accuracy: {accuracy * 100:.2f}%")
print(f"F1 Score: {f1:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print("Confusion Matrix:")
print(conf_matrix)
```
